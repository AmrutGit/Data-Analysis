<h2>Time series analysis all models with their use

<h3>1. Moving Average (MA) Models</h3>
<ul>
  <li><b>Simple Moving Average (SMA)</b>: The simplest time series model, which averages the data over a specified period.</li>
  <li><b>Weighted Moving Average (WMA)</b>: Similar to SMA but with different weights applied to the data points.</li>
  <li><b>Exponential Moving Average (EMA)</b>: A variation of the moving average that gives more weight to recent observations, making it more sensitive to new data points.</li>
</ul>

<h3>2. Autoregressive Integrated Moving Average (ARIMA) Models</h3>
<ul>
  <li><b>ARIMA (AutoRegressive Integrated Moving Average)</b>: One of the most popular models for time series forecasting, combining autoregressive (AR) terms, differencing (I) for stationarity, and moving averages (MA).</li>
  <li><b>SARIMA (Seasonal ARIMA)</b>: Extends ARIMA by including seasonal components to handle seasonal patterns in time series data.</li>
  <li><b>ARIMAX (ARIMA with exogenous variables)</b>: Incorporates external factors or regressors (like interest rates, inflation, etc.) to improve forecasts.</li>
</ul>

<h3>3. Autoregressive (AR) Models</h3>
<ul>
  <li><b>AR(p) Model</b>: Focuses only on autoregressive terms where the current value is based on its past values (lags). It's simpler than ARIMA and useful for non-stationary time series with autocorrelation.</li>
</ul>

<h3>4. Moving Average (MA) Models</h3>
<ul>
  <li><b>MA(q) Model</b>: Focuses only on the moving average terms, where the current value is a function of past error terms (noise).</li>
</ul>

<h3>5. GARCH Models (Generalized Autoregressive Conditional Heteroskedasticity)</h3>
<ul>
  <li><b>ARCH (Autoregressive Conditional Heteroskedasticity)</b>: Models time-varying volatility and is often used in modeling stock returns, currency exchange rates, and financial market volatility.</li>
  <li><b>GARCH (Generalized ARCH)</b>: An extension of the ARCH model that includes lagged forecast errors and is more flexible in capturing volatility clustering.</li>
  <li><b>EGARCH (Exponential GARCH)</b>: Accounts for asymmetric volatility, where negative shocks might have a greater impact than positive shocks.</li>
</ul>

<h3>6. Vector Autoregression (VAR) Models</h3>
<ul>
  <li><b>VAR Model</b>: A multivariate time series model that captures the linear relationships among multiple time series variables. It is often used in macroeconomic forecasting, such as GDP, interest rates, and inflation.</li>
</ul>

<h3>7. Cointegration Models</h3>
<ul>
  <li><b>Engle-Granger Two-Step Cointegration</b>: Used when there is a long-term relationship between two or more non-stationary time series.</li>
  <li><b>Johansen Cointegration Test</b>: A more sophisticated test for cointegration among multiple time series.</li>
  <li><b>Vector Error Correction Model (VECM)</b>: A model that combines VAR and cointegration, useful when the variables are cointegrated and have long-term equilibrium relationships.</li>
</ul>

<h3>8. State-Space Models</h3>
<ul>
  <li><b>Kalman Filter</b>: A recursive algorithm that estimates the hidden states of a system from noisy observations. It's used in financial modeling for state estimation and forecasting.</li>
  <li><b>Dynamic Linear Models (DLM)</b>: A broad class of state-space models useful for time series forecasting where the system's state evolves over time.</li>
</ul>

<h3>9. Long Short-Term Memory (LSTM) Networks</h3>
<ul>
  <li><b>LSTM Networks</b>: A type of recurrent neural network (RNN) designed to learn and predict time-dependent data. It has been widely used in stock price prediction, crypto analysis, and other financial market modeling due to its ability to learn long-term dependencies in data.</li>
</ul>

<h3>10. Prophet</h3>
<ul>
  <li><b>Prophet</b>: A forecasting tool developed by Facebook that is designed to handle seasonality and trends. It's easy to use and is effective for daily or weekly data with seasonal patterns.</li>
  <li><b>Prophet with Holidays</b>: In addition to daily and weekly seasonality, Prophet can incorporate public holidays and special events, which may have an impact on stock price movements.</li>
</ul>

<h3>11. Neural Networks for Time Series</h3>
<ul>
  <li><b>Feedforward Neural Networks (FNN)</b>: Basic neural network models that can be applied to time series forecasting.</li>
  <li><b>Convolutional Neural Networks (CNN)</b>: Used in financial time series forecasting by treating the time series data as 1D images, which can capture patterns effectively.</li>
  <li><b>Recurrent Neural Networks (RNN)</b>: RNNs, including variants like GRU and LSTM, are useful for sequential data like time series forecasting.</li>
</ul>

<h3>12. Support Vector Machines (SVM) for Regression</h3>
<ul>
  <li><b>SVR (Support Vector Regression)</b>: A method for predicting a continuous value, such as stock prices, by mapping input data into a high-dimensional space and finding the best-fitting hyperplane.</li>
  <li><b>SVR with Time Series Data</b>: SVR can be used for forecasting financial time series by selecting appropriate kernels.</li>
</ul>

<h3>13. Random Forests and Gradient Boosting</h3>
<ul>
  <li><b>Random Forest</b>: A machine learning algorithm that uses an ensemble of decision trees to model time series data. It's useful for predicting financial outcomes with complex patterns.</li>
  <li><b>XGBoost (Extreme Gradient Boosting)</b>: An advanced boosting algorithm that works well with time series data by capturing non-linear patterns and interactions between variables.</li>
</ul>

<h3>14. Bayesian Methods</h3>
<ul>
  <li><b>Bayesian Structural Time Series (BSTS)</b>: A probabilistic model for time series analysis that uses Bayesian inference to model trend, seasonality, and irregular components of a time series.</li>
  <li><b>Bayesian Networks</b>: Graphical models that can model the conditional dependencies between different financial time series, useful in forecasting stock prices.</li>
</ul>

<h3>15. Monte Carlo Simulation</h3>
<ul>
  <li><b>Monte Carlo Methods</b>: Use random sampling to simulate the behavior of a financial time series and to forecast future values. It's often used to estimate risk and uncertainty in financial models.</li>
  <li><b>Simulating Stock Prices</b>: Monte Carlo is widely used for simulating future stock prices using stochastic models like the Geometric Brownian Motion (GBM).</li>
</ul>

<h3>16. Kalman Filter Models</h3>
<ul>
  <li><b>Extended Kalman Filter (EKF)</b>: A non-linear version of the Kalman filter used when the system model is non-linear, often applied in financial forecasting when the underlying system is not purely linear.</li>
</ul>

<h3>17. Stochastic Differential Equations (SDE)</h3>
<ul>
  <li><b>Geometric Brownian Motion (GBM)</b>: A stochastic process used to model stock prices and asset returns, where the rate of return follows a log-normal distribution.</li>
  <li><b>Mean Reversion Models</b>: Models like the Ornstein-Uhlenbeck process are used in financial time series to predict prices that tend to revert to a long-term mean.</li>
</ul>

<h3>Choosing the Right Model</h3>
<p>
  The choice of model depends on various factors:
</p>
<ul>
  <li><b>Data Stationarity</b>: If your data is non-stationary (i.e., it has trends or seasonality), methods like ARIMA, SARIMA, or Prophet are more appropriate.</li>
  <li><b>Volatility</b>: Models like GARCH and its variants are specifically designed for volatility modeling, common in financial markets.</li>
  <li><b>Seasonality</b>: If the time series exhibits strong seasonal patterns, Prophet, SARIMA, or even machine learning models that can handle periodicity (e.g., Random Forest, XGBoost) can work well.</li>
  <li><b>Complexity and Interpretability</b>: ARIMA and GARCH models are well-understood, interpretable models, while machine learning models like LSTM or XGBoost can model complex relationships but are harder to interpret.</li>
</ul>

```python
# Import necessary libraries
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.arima.model import ARIMA
from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error
import arch
from statsmodels.tsa.api import VAR
from statsmodels.tsa.stattools import coint, adfuller
from pykalman import KalmanFilter
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
# Download stock data
def download_stock_data(ticker, start_date, end_date):
    data = yf.download(ticker, start=start_date, end=end_date)
    data.reset_index(inplace=True)
    return data
```

```python
# Load data for Tesla and Google
start_date = "2023-12-01"
end_date = "2024-12-01"

tesla_data = download_stock_data("TSLA", start_date, end_date)
google_data = download_stock_data("GOOGL", start_date, end_date)
```

```python
# Helper function to plot time series data
def plot_stock_data(data, title):
    plt.figure(figsize=(12, 6))
    plt.plot(data['Date'], data['Close'], label='Close Price')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
# Plot data
plot_stock_data(tesla_data, "Tesla Stock Price")
plot_stock_data(google_data, "Google Stock Price")
```

<h3> Functions to evaluate models

```python
# Function to evaluate models
def evaluate_model(true, predicted, model_name, stock_name):
    mae = mean_absolute_error(true, predicted)
    mse = mean_squared_error(true, predicted)
    print(f"{stock_name} - {model_name}: MAE={mae:.2f}, MSE={mse:.2f}")
    return mae, mse
```

```python
# Exponential Smoothing Model
def exponential_smoothing_model(data, stock_name):
    train = data['Close'][:-30]
    test = data['Close'][-30:]
    model = ExponentialSmoothing(train, seasonal='add', seasonal_periods=12).fit()
    forecast = model.forecast(30)
    evaluate_model(test, forecast, "Exponential Smoothing", stock_name)
    plt.figure(figsize=(12, 6))
    plt.plot(data['Date'][-60:], data['Close'][-60:], label='True')
    plt.plot(data['Date'][-30:], forecast, label='Forecast')
    plt.title(f"{stock_name} - Exponential Smoothing Forecast")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
# ARIMA Model
def arima_model(data, stock_name):
    train = data['Close'][:-30]
    test = data['Close'][-30:]
    model = ARIMA(train, order=(5, 1, 0)).fit()
    forecast = model.forecast(30)
    evaluate_model(test, forecast, "ARIMA", stock_name)
    plt.figure(figsize=(12, 6))
    plt.plot(data['Date'][-60:], data['Close'][-60:], label='True')
    plt.plot(data['Date'][-30:], forecast, label='Forecast')
    plt.title(f"{stock_name} - ARIMA Forecast")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
# Prophet Model
def prophet_model(data, stock_name, tic):
    # Extract relevant columns using the MultiIndex
    df = data[('Date', '')].to_frame()  # Extract 'Date' from MultiIndex
    df['y'] = data[('Close', tic)]  # Extract 'Close' for 'y'
    
    # Rename columns to meet Prophet's requirements
    df.columns = ['ds', 'y']
    
    # Ensure that 'ds' (Date) has a consistent frequency
    df['ds'] = pd.to_datetime(df['ds'])  # Convert to datetime if not already
    df = df.set_index('ds').sort_index()  # Set the 'ds' column as index
    
    # Fill missing dates with linear interpolation (if needed)
    df = df.asfreq('D', method='pad')  # Ensure daily frequency

    # Prepare the training and test sets
    train = df.iloc[:-30]
    test = df.iloc[-30:]

    # Reset index to ensure 'ds' is a column, not the index
    train = train.reset_index()  # Now 'ds' is a column

    # Ensure the train DataFrame has the correct structure
    # print(f"Train DataFrame columns: {train.columns}")
    # print(f"Train DataFrame head:\n{train.head()}")

    # Fit Prophet model
    model = Prophet()
    model.fit(train)
    
    # Make predictions
    future = model.make_future_dataframe(periods=30, freq='D')  # Generate future dates manually
    forecast = model.predict(future)
    
    # Evaluate the model
    predicted = forecast['yhat'].iloc[-30:]
    evaluate_model(test['y'], predicted, "Prophet", stock_name)
    
    # Plot results
    plt.figure(figsize=(12, 6))
    plt.plot(df.index, df['y'], label='True')
    plt.plot(forecast['ds'], forecast['yhat'], label='Forecast')
    plt.title(f"{stock_name} - Prophet Forecast")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
print("Tesla Analysis:")
exponential_smoothing_model(tesla_data, "Tesla")
arima_model(tesla_data, "Tesla")
prophet_model(tesla_data, "Tesla", 'TSLA')
```

```python
print("\nGoogle Analysis:")
exponential_smoothing_model(google_data, "Google")
arima_model(google_data, "Google")
prophet_model(google_data, "Google", 'GOOGL')
```

<h2> Dynamic Linear Model

```python
# Dynamic Linear Model using SARIMAX with evaluation
def dlm_model(data, stock_name):
    # Split data into train and test (last 30 days as test)
    train = data['Close'][:-30]
    test = data['Close'][-30:]
    
    # SARIMAX model fitting (Dynamic Linear Model)
    model = SARIMAX(train, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
    model_fit = model.fit()
    
    # Forecast for the next 30 days
    forecast = model_fit.forecast(steps=30)
    
    # Evaluate the model on the test set
    evaluate_model(test, forecast, "Dynamic Linear Model (SARIMAX)", stock_name)
    
    # Plot the true vs predicted values
    plt.figure(figsize=(12, 6))
    plt.plot(data['Date'][-60:], data['Close'][-60:], label='True')
    plt.plot(data['Date'][-30:], forecast, label='Forecast', linestyle='--')
    plt.title(f"{stock_name} - Dynamic Linear Model (SARIMAX) Forecast")
    plt.legend()
    plt.grid()
    plt.show()
```

<h3>ARCH, GARCH, and EGARCH Models</h3>

```python
# Import necessary libraries for new models
from arch import arch_model
from statsmodels.tsa.api import VAR
from statsmodels.tsa.vector_ar.vecm import coint_johansen
from pykalman import KalmanFilter
import statsmodels.api as sm
```

```python
# GARCH Models with Complete Time Frame Plot (Corrected)
def garch_model(data, stock_name):
    returns = 100 * data['Close'].pct_change().dropna()
    train = returns[:-30]
    test = returns[-30:]
    
    # Fit GARCH(1,1) model
    model = arch_model(train, vol='Garch', p=1, q=1)
    model_fitted = model.fit(disp="off")
    forecast = model_fitted.forecast(horizon=30)
    
    # Extract forecasted volatility and flatten it
    forecasted_volatility = forecast.variance.iloc[-1].to_numpy().flatten()
    
    # Evaluate the model
    evaluate_model(test, forecasted_volatility, "GARCH", stock_name)
    
    # Plot volatility forecast
    plt.figure(figsize=(12, 6))
    
    # Plot the complete time frame
    plt.plot(data['Date'][1:], returns, label='Returns (Complete Time Frame)', alpha=0.7)
    
    # Highlight the forecasted period
    forecast_dates = data['Date'][-30:]
    plt.plot(forecast_dates, forecasted_volatility, label='Forecast Volatility', color='orange')
    
    plt.title(f"{stock_name} - GARCH Model (Complete Time Frame)")
    plt.xlabel("Date")
    plt.ylabel("Volatility (%)")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
print("GARCH Model:")
garch_model(tesla_data, "Tesla")
garch_model(google_data, "Google")
```

<h1>Vector Autoregression (VAR) Models</h1>

<p>
  <b>VAR Model</b>: A multivariate time series model that captures the linear relationships among multiple time series variables. It is often used in macroeconomic forecasting, such as GDP, interest rates, and inflation.
</p>


```python
from statsmodels.tsa.api import VAR

def var_model(data, stock_name):
    # Selecting relevant columns for VAR
    variables = data[['Close', 'Volume']]  # Ensure you have more than one column
    returns = 100 * variables.pct_change().dropna()
    
    train = returns[:-30]
    test = returns[-30:]
    
    # Fit VAR model
    model = VAR(train)
    model_fitted = model.fit(5)  # Use 5 lags
    forecast = model_fitted.forecast(train.values[-5:], steps=30)
    
    # Evaluate and Plot
    evaluate_model(test, forecast, "VAR", stock_name)
    plot_full_time_frame(data, forecast, stock_name, "VAR")
```

```python
print("\nTesla VAR Model:")
var_model(tesla_data, "Tesla")
var_model(google_data, "Google")
```

<h1>Kalman Filter</h1>

<p>
  <b>Kalman Filter</b>: A recursive algorithm that estimates the hidden states of a system from noisy observations. It's used in financial modeling for state estimation and forecasting.
</p>

```python
def kalman_filter_model(data, stock_name):
    kf = KalmanFilter(initial_state_mean=0, n_dim_obs=1)
    kf = kf.em(data['Close'], n_iter=10)
    
    # Predict for the entire series
    (filtered_state_means, _) = kf.filter(data['Close'])
    forecast = filtered_state_means[-30:]
    
    # Evaluate and Plot
    evaluate_model(data['Close'][-30:], forecast, "Kalman Filter", stock_name)
    plot_full_time_frame(data, filtered_state_means, stock_name, "Kalman Filter")

```

```python
print("\nTesla Kalman Filter Model:")
kalman_filter_model(tesla_data, "Tesla")
kalman_filter_model(google_data, "Google")
```

<h1>Cointegration Models</h1>

<ul>
  <li>
    <b>Engle-Granger Two-Step Cointegration</b>: Used when there is a long-term relationship between two or more non-stationary time series.
  </li>
  <li>
    <b>Johansen Cointegration Test</b>: A more sophisticated test for cointegration among multiple time series.
  </li>
  <li>
    <b>Vector Error Correction Model (VECM)</b>: A model that combines VAR and cointegration, useful when the variables are cointegrated and have long-term equilibrium relationships.
  </li>
</ul>

<p>
  <b>Note</b>: Although there is no relationship between Tesla stock and Google stock, we are doing this analysis just for demonstration purposes.
</p>


```python
def engle_granger_cointegration(data1, data2, stock_name1, stock_name2):
    returns1 = 100 * data1['Close'].pct_change().dropna()
    returns2 = 100 * data2['Close'].pct_change().dropna()
    
    # Align lengths of returns
    min_length = min(len(returns1), len(returns2))
    returns1 = returns1[-min_length:]
    returns2 = returns2[-min_length:]
    
    # Perform Engle-Granger cointegration test
    score, p_value, _ = coint(returns1, returns2)
    
    # Print cointegration result
    print(f"{stock_name1} & {stock_name2} Cointegration Test: p-value = {p_value:.5f}")
    if p_value < 0.05:
        print(f"The time series {stock_name1} and {stock_name2} are cointegrated.")
    else:
        print(f"The time series {stock_name1} and {stock_name2} are NOT cointegrated.")
    
    # Plot the time series and their residuals
    residuals = returns1 - returns2
    plot_cointegration_time_series(returns1, returns2, residuals, stock_name1, stock_name2)

def plot_cointegration_time_series(returns1, returns2, residuals, stock_name1, stock_name2):
    plt.figure(figsize=(14, 8))
    
    # Plot the two time series
    plt.subplot(2, 1, 1)
    plt.plot(returns1.index, returns1, label=f"{stock_name1} Returns", color="blue")
    plt.plot(returns2.index, returns2, label=f"{stock_name2} Returns", color="orange")
    plt.title(f"Cointegration Analysis: {stock_name1} and {stock_name2}")
    plt.xlabel("Time")
    plt.ylabel("Returns")
    plt.legend()
    plt.grid()
    
    # Plot the residuals
    plt.subplot(2, 1, 2)
    plt.plot(residuals.index, residuals, label="Residuals", color="green")
    plt.axhline(0, color="red", linestyle="--", label="Equilibrium Line")
    plt.title("Residuals of Cointegration")
    plt.xlabel("Time")
    plt.ylabel("Residuals")
    plt.legend()
    plt.grid()
    
    plt.tight_layout()
    plt.show()
```

```python
print("\nTesla Engle-Granger Cointegration (with Google):")
engle_granger_cointegration(tesla_data, google_data, "Tesla", "Google")
```

```python
def johansen_cointegration(data, stock_name):
    # Select multiple columns for multivariate analysis
    variables = data[['Close', 'Volume']]  # Example with 'Close' and 'Volume'
    variables = 100 * variables.pct_change().dropna()  # Calculate percentage change
    
    # Perform Johansen Cointegration Test
    johansen_test = coint_johansen(variables.values, det_order=0, k_ar_diff=1)
    
    # Print cointegration test results
    print(f"{stock_name} Johansen Cointegration Test Trace Statistic: {johansen_test.lr1}")
    print(f"{stock_name} Johansen Cointegration Test Critical Value: {johansen_test.cvt}")
    
    # Plot the first column of returns
    plt.figure(figsize=(12, 6))
    plt.plot(data['Date'][-60:], variables.iloc[-60:, 0], label='Returns (Close)')
    plt.title(f"{stock_name} - Johansen Cointegration Test")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
print("\nTesla Johansen Cointegration Test:")
johansen_cointegration(tesla_data, "Tesla")
johansen_cointegration(google_data, "Google")
```

```python
from statsmodels.tsa.api import VECM

def vecm_model(data1, data2, stock_name1, stock_name2):
    returns1 = 100 * data1['Close'].pct_change().dropna()
    returns2 = 100 * data2['Close'].pct_change().dropna()
    
    # Prepare data for VECM (Vector Error Correction Model)
    data = pd.concat([returns1, returns2], axis=1).dropna()
    model = VECM(data, k_ar_diff=1)
    model_fitted = model.fit()
    forecast = model_fitted.predict(steps=30)
    
    # Evaluate the model
    evaluate_model(returns2[-30:], forecast[:, 1], "VECM", f"{stock_name1} & {stock_name2}")
    
    # Generate future dates for the forecast
    last_date = data2['Date'].iloc[-1]
    future_dates = pd.date_range(start=last_date, periods=30, freq='B')  # Adjust frequency as needed
    
    # Plot the forecasted VECM returns
    plt.figure(figsize=(12, 6))
    plt.plot(data1['Date'][-60:], returns1[-60:], label=f'{stock_name1} Returns')
    plt.plot(data2['Date'][-60:], returns2[-60:], label=f'{stock_name2} Returns')
    plt.plot(future_dates, forecast[:, 1], label=f'VECM Forecast for {stock_name2}', linestyle='--')
    plt.title(f"{stock_name1} & {stock_name2} - VECM Forecast")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
print("\nTesla VECM Model (with Google):")
vecm_model(tesla_data, google_data, "Tesla", "Google")
```
